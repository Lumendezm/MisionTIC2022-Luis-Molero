```python
# CAMILO 
import json
import pandas as pd

def funcion(lista_1, lista_2):
    # creamos un dataframe a partir de dos listas
    
    diccionario_lista = {
        'lista_1': lista_1,
        'lista_2': lista_2,
    }

    datosdataframe = pd.DataFrame(diccionario_lista)
    print(datosdataframe)

    # Creando el archivo csv
    datosdataframe.to_csv('ejercicio.csv')
    
    # Creando el archivo JSON
    with open('ejercicio.json', 'w') as file:
        json.dump(datosdataframe, file, indent=4)
    
# creamos dos listas que serán los argumentos de entrada
Lista1 = [1,2,3]  
Lista2 = [4,5,6]

# Llamamos la función "funcion"
funcion(Lista1,Lista2)
```
